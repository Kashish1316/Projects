```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv(r'C:\Users\Kashi\Downloads\book.csv')
```

```python
books.head()
```

```python
movies = pd.read_csv(r'C:\Users\Kashi\Downloads\my_movies.csv')
```

```python
movies.head()
```

```python
frequent_itemsets = apriori(books, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support', ascending = False)
```

```python
rules = association_rules(frequent_itemsets, metric = 'confidence')
rules
```

```python
rules.sort_values('lift', ascending = False)
```

```python
rules[rules.lift<1]
```

```python
# Example: Frequent itemsets with support >= 0.1 and confidence >= 0.5
rules1 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules1
```

```python
# Filter rules with a minimum of 2 items in antecedents and consequents
filtered_rules = rules[(rules['antecedents'].apply(len) >= 2) & (rules['consequents'].apply(len) >= 1)]
filtered_rules
```

```python
plt.scatter(rules1['support'], rules1['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs. Confidence')
plt.show()
```

```python
plt.scatter(rules1['confidence'], rules1['lift'], alpha=0.5)
plt.xlabel('Confidence')
plt.ylabel('Lift')
plt.title('Confidence vs. Lift')
plt.show()
```

```python
plt.scatter(rules1['support'], rules1['lift'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Lift')
plt.title('Support vs. Lift')
plt.show()
```
